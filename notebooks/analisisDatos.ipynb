```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde el archivo de entrada
archivo_entrada = "../data/processed/productosProcesados.csv"
if archivo_entrada.endswith(".csv"):
    df = pd.read_csv(archivo_entrada)
elif archivo_entrada.endswith(".xlsx"):
    df = pd.read_excel(archivo_entrada)

# Análisis Estadístico Básico
print("Análisis Estadítico Básico: ")
estadistica_basica = df.describe().map(lambda x: f'{x:.2f}')
print(estadistica_basica)

# Histograma de la distribución de precios
plt.hist(df["Precio"], bins=20, edgecolor="black")
plt.title("Distribución de Precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

# Productos con los precios más altos
print("Los 5 productos con los precios más altos: ")
highest_products=df.nlargest(5, "Precio")
print(highest_products)

# Productos con los precios más bajos
print("Los 5 productos con los precios más bajos: ")
smallest_products=df.nsmallest(5, 'Precio')
print(smallest_products)

# Calcular la frecuencia de cada marca
conteo_marcas = df['Marca'].value_counts()

# Gráfico de barras de la oferta de cada marca
plt.figure(figsize=(7, 4))
conteo_marcas.plot(kind='bar', color=['blue', 'green', 'red', 'orange', 'purple'])
for x, y in enumerate(conteo_marcas):
    plt.text(x, y + 0.2, str(y), ha='center', va='bottom')
plt.title('Oferta de cada marca')
plt.xlabel('Marca')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de pastel de la distribución de marcas
plt.figure(figsize=(5,5))
plt.pie(conteo_marcas, labels=conteo_marcas.index, autopct='%1.1f%%', startangle=225)
plt.title('Distribución de la oferta por marcas')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
